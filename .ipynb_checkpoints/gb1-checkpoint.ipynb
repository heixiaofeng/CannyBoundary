```python
from PIL import Image

from pylab import *

# 读取图像到数组中 

# convert('L')表示将RGB转换为L模式，表示像素点在[0,255]之间

im = array(Image.open('img/cat.jpg').convert('L'))

# 新建一个图像

figure()

# 使用颜色信息

gray()

# 在原点的左上角显示轮廓图像

contour(im, levels=[90],colors='black',origin='image')

axis('equal')

axis('off')

figure()

hist(im.flatten(),128)

show()
```
